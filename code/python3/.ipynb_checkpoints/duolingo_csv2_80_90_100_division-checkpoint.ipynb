```python
import pandas as pd 


def combine_train_valid_test_sheets(path):
    train_df=pd.DataFrame.from_csv(path)
    valid_df=pd.DataFrame.from_csv(path.replace("train","dev"))
    test_df=pd.DataFrame.from_csv(path.replace("train","test"))
    print(len(train_df.columns))
    print(train_df.columns)
    print(len(valid_df.columns))
    print(len(test_df.columns))
    print(set(train_df.columns).difference(set(valid_df.columns)))
    print(set(train_df.columns).difference(set(test_df.columns)))
    frames = [train_df, valid_df, test_df]
    result = pd.concat(frames,keys=['train','dev','test'],names=['from', 'idx'])
    return result
    
result=combine_train_valid_test_sheets("../../data/duolingo/es_en.slam.20171218.train.csv")

```

```python
concat_result=result.reset_index(level=['from']).reset_index(drop=True)
from collections import defaultdict
user_list=list(concat_result['user'].unique())
# print(user_list)

append=True
directory="../../data/duolingo/new_split_append"+str(append)+"_"

user2id=defaultdict(lambda:len(user2id))
with open(directory+"user_id_mapping2.txt","w") as f:
    for user in user_list:
        f.write(user+"\t"+str(user2id[user])+"\n")
    
selected_problem_name="token"




user_history=concat_result[['from','token','part_of_speech','correctness','user']].groupby(['user']).agg({'token':lambda x: tuple(x), 
                                   'part_of_speech':lambda x: tuple(x),'correctness':lambda x:tuple(x),'from':lambda x:tuple(x)})


problem2id=defaultdict(lambda:len(problem2id)+1)

f_train=open(directory+selected_problem_name+"_unsplitted_train1.csv","w")
f_valid=open(directory+selected_problem_name+"_unsplitted_valid1.csv","w")
f_test=open(directory+selected_problem_name+"_unsplitted_test.csv","w")
f_user_division=open(directory+selected_problem_name+"_unsplitted_user_division","w")
for idx,row in user_history.iterrows():
    from_list=list(row['from'])
    train_len=len([value for value in from_list if value=="train"])
    dev_len=len([value for value in from_list if value=="dev"])
    test_len=len([value for value in from_list if value=="test"])
    
    assert (train_len+dev_len+test_len)==len(from_list)
    
    problem_list=list(row[selected_problem_name])
    problem_id_list=[str(problem2id[item]) for item in problem_list]
    correctness_list=[str(int(value)) for value in list(row['correctness'])]
    
    cur_len=train_len
    f_train.write(str(cur_len)+" # "+str(user2id[idx])+"\n"+str(",".join(problem_id_list[:cur_len]))+"\n"+",".join(correctness_list[:cur_len])+"\n")
    prev_len=0
    if not append:
        prev_len=train_len
    cur_len=train_len+dev_len
    f_valid.write(str(cur_len-prev_len)+" # "+str(user2id[idx])+"\n"+str(",".join(problem_id_list[prev_len:cur_len]))+"\n"+",".join(correctness_list[prev_len:cur_len])+"\n")

    if not append:
        prev_len=train_len+dev_len
    cur_len=train_len+dev_len+test_len
    f_test.write(str(cur_len-prev_len)+" # "+str(user2id[idx])+"\n"+str(",".join(problem_id_list[prev_len:cur_len]))+"\n"+",".join(correctness_list[prev_len:cur_len])+"\n")
    
    f_user_division.write(str(user2id[idx])+"\t"+str(train_len)+"\t"+str(dev_len)+"\t"+str(test_len)+"\n")
    
with open(directory+selected_problem_name+"_unsplitted_id_mapping.txt","w") as f:
    for problem in problem2id:
        f.write(problem+"\t"+str(problem2id[problem])+"\n")
        
    
    
    
```
