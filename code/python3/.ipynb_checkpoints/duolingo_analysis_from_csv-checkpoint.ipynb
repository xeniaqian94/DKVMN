```python
user_seq=(df[['token','part_of_speech','correctness','user']].groupby(['user']).agg({'token':lambda x: tuple(x), 
                                   'part_of_speech':lambda x: tuple(x),'correctness':lambda x:tuple(x)}))
```

```python
selected_question_skill_colum="token"
question2id=defaultdict(lambda:len(question2id))

with open(dir_path+selected_question_skill_colum+"_train.txt","w") as f:
    for question in question2id:
        f.write(question+"\t"+str(question2id[question])+"\n")
        
for idx,row in user_seq.iterrows():
    problems=list(row[selected_question_skill_colum])
    problems_id=[str(question2id[problem]) for problem in problems]
    correctness_int=[str(int(value)) for value in list(row['correctness'])]
    print(str(len(problems))+" # "+str(user2id[idx])+"\n"+",".join(problems_id)+"\n"+",".join(correctness_int))

with open(dir_path+selected_question_skill_colum+"_id_mapping.txt","w") as f:
    for question in question2id:
        f.write(question+"\t"+str(question2id[question])+"\n")
```
